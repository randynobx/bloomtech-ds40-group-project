```python
'''
Basic data cleaning notebook
'''
import sqlite3
import pandas as pd
import sys
if '..' not in sys.path:
    sys.path.insert(0, '..')
from src.loaders.config import Config

# Create config object
config = Config('../config')

# make a sql connection
# use temporary db
config.database = '../data/boardgames.sqlite3'
sql_db = sqlite3.connect(config.database)
```

```python
def is_everything_imported(df, table: str, db):
    '''
    Checking is number of rows in dataframe == number of rows in SQL table
    Return True or False
    '''
    cursor = db.cursor()
    num_of_cols_in_sql_q = f"""
    SELECT COUNT(*)
    FROM pragma_table_info('{table}');
    """
    num_of_rows_in_sql_q = f"""
    SELECT COUNT(*)
    FROM {table};
    """
    num_of_cols_in_sql = cursor.execute(num_of_cols_in_sql_q).fetchall()[0][0]
    num_of_rows_in_sql = cursor.execute(num_of_rows_in_sql_q).fetchall()[0][0]  
    return df.shape == (num_of_rows_in_sql, num_of_cols_in_sql)


def read_sql_table_to_df(table: str, db):
    '''
    Read table from SQL db to pandas Dataframe
    '''
    df = pd.read_sql_query(f'SELECT * FROM {table}', db)
    if not is_everything_imported(df, table, db):
        raise ValueError ('Error while importing sql to dataframes, rows in dataframs != rows in sql table')
    return df
```

```python
# Reading all sql tables to dataframes
df_game = read_sql_table_to_df('game', sql_db)
df_category = read_sql_table_to_df('category', sql_db)
df_game_category_map = read_sql_table_to_df('game_category_map', sql_db)
df_game_mechanic_map = read_sql_table_to_df('game_mechanic_map', sql_db)
df_mechanic = read_sql_table_to_df('mechanic', sql_db)
```
